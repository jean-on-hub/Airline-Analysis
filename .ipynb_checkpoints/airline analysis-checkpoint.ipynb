### import dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
## Read in Data
flight = pd.read_csv("flight.csv")
flight.head()
```

```python
## Task 1
# analysing the price data
sns.kdeplot(flight.coach_price,shade =0.2)
print(flight.coach_price.mean())
print(flight.coach_price.median())
#  from the plot we can see that 500$  is not an average price and is closer to the tail of the data and therefore not an appropriate price for  a coach ticket. the median and mean also show that 500 is nowhere near an approriate price
```

```python
## Task 2
# seperating 8hrs flights
flight8hrs = flight[flight["hours"]==8]
# print(flight8hrs)
sns.kdeplot(flight.coach_price,shade =0.2)
print(flight8hrs.coach_price.mean())
print(flight8hrs.coach_price.median())

# from the plot we can see that is not an average price and is closer to the tail of the data and therefore not an appropriate price for  a coach ticket. the median and mean also show that 500 is nowhere near an approriate price
```

```python
## Task 3
# print(flight.delay.unique())
# removing outliers
flight_no_outlier = flight[flight["delay"]<20]
sns.kdeplot(flight_no_outlier.delay,shade =0.2)
print(flight_no_outlier.delay.mean())
print(flight_no_outlier.delay.median())


# from the data even though there are obvious outliers the average delay time is around 10 mins so preparing for a 10 min delay is most aprropriate
```

```python
## Task 4
# relationship between coach and first class prices

# use a 20% sample as the dataset is big
perc =0.2
flight_sample = flight.sample(n = int(flight.shape[0]*perc))
sns.lmplot(x = "coach_price", y = "firstclass_price", data = flight_sample, line_kws={'color': 'black'}, lowess=True)
# plt.scatter(flight_sample.firstclass_price,flight_sample.coach_price)

# from the line of the best fit we can confidently say there's a positive relationship between coach and first class prices
# the lmplot shows
```

```python
## Task 5
sns.set_palette("Set3")
sns.set_context("paper")
plt.figure(figsize=(15, 10))
plt.subplot(1,3,1)
sns.boxplot(data = flight,x ="inflight_meal", y = "coach_price")
plt.subplot(1,3,2)
sns.boxplot(data = flight,x ="inflight_entertainment", y = "coach_price")
plt.subplot(1,3,3)
sns.boxplot(data = flight,x ="inflight_wifi", y = "coach_price")
# plt.subplots_adjust(left = 0.5,right = 0.5)

# from the sublplots it is easy to see that inflight features raise coach prices considerably but inflight entertainment tops the three when it comes to raising the price
```

```python
## Task 6
perc =0.05
flight_sample = flight.sample(n = int(flight.shape[0]*perc))
sns.lmplot(x = "hours", y = "passengers", data = flight_sample, scatter_kws={"alpha" : 0.5, "s":5},x_jitter= 0.1,y_jitter = 0.1, lowess = False,fit_reg= False)
# plt.show()
plt.clf()
# from the above figure it can be seen that the data is seperated in  chuncks of above 180 and below 180 so i will seperate them an anylise them together
flight_above = flight[flight.passengers >= 180]
flight_sample = flight_above.sample(n = int(flight_above.shape[0]*perc))
sns.lmplot(x = "hours", y = "passengers", data = flight_sample, scatter_kws={"alpha" : 0.5, "s":5},x_jitter= 0.1,y_jitter = 0.1, lowess = False,fit_reg= True)
# plt.show()
plt.clf()
flight_below = flight[flight.passengers < 180]
perc =0.5
flight_sample = flight_below.sample(n = int(flight_below.shape[0]*perc))
sns.lmplot(x = "hours", y = "passengers", data = flight_sample, scatter_kws={"alpha" : 0.5, "s":5},x_jitter= 0.1,y_jitter = 0.1, lowess = False,fit_reg= True)

# from the 2 plots there's no obvious relationship between number of passengers and flight hours even though the number of passengers obviously dimunishes in 6 and 8 hour flights
```

```python
## Task 7
# use a 20% sample as the dataset is big
perc =0.2
flight_sample = flight.sample(n = int(flight.shape[0]*perc))
sns.lmplot(x = "coach_price", y = "firstclass_price", data = flight_sample, line_kws={'color': 'black'}, lowess=True,hue = "weekend",x_jitter= 0.1)
# plt.show()
plt.clf()
# from the data flight tickests are generally less expensive on weekdays  
```
